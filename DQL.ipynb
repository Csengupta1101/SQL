As the name suggests , this type of queries are for querying into the data. In all likelihood , the very first layer in our query is generally the DQL or data query language. Or in other words it is used to retireve the information from the table.

The 'SELECT' Command is the example of DQL .When we want to retrieve the data in it's actual form or we want to display the entire table , we use "SELECT * FROM table_name" as below. In here we are displaying the employees data in a company data base table.

    -   SELECT * FROM employee;

Now suppose we want to display only the first_name of the employees from the table , then we will use the syntax as below

    -   SELECT first_name FROM employee;

We can also select multiple columns based on our requirements here , suppose we want to select only first name and the last name , we can use the syntax as below 

    -   SELECT first_name , last_name FROM employee;

Now there are few clauses that we can provide as an additional input with the select command for filetring the datya based on our specific condition. This is called conditional fltering. The most commonly used clauses are as below ,

    -   WHERE
    -   ORDER BY
    -   GROUP BY
    -   HAVING

Let's talk about where clause first. As the name suggests , this simply identifies the specific requirement we have from the data. Suppose from our employee database we want to select only the male employees and we have seen that under 'sex' column gender is specified as 'm' or 'f' so we can run a where clause as below.

    -   SELECT * FROM employee WHERE sex ='m';

We can also use mathematical condition over here , suppose we want to only display employee details where the birth dates are before 1967. So ideally we will use a where clause stating that we need all employee data who were born before 1st january 1967. The query will look like as below ,

    -   SELECT * FROM employee WHERE birth_day < '1967-01-01' ;

As the name suggests , order by clause is used to retrieve the data in a given specific order, Suppose we want to retrieve all data sorted by the employee salaries , the query will be like as below ,

    -   SELECT * FROM employee ORDER BY salary;

If we want the data to be displayed in high to low salary order then we simply need to add DESC at the end of the query which will reflect our intent to display the result in the high-low order.

    -   SELECT * FROM employee ORDER BY salary DESC;

To use group by statement in your select command reflects that we want to group by our data based on a given condition. This simply means that group by works on a given condition , so 'WHERE' clause will be required to execute a group by statement. Suppose we want to group by the first name and last name based on gender and the condition will be salary being less than 200000 , and find out the count for each category we can use a syntax as below

    -   SELECT count(first_name) FROM employee WHERE salary < 200000 GROUP BY sex;

'HAVING' clause and 'WHERE' clause works almost similarly with 'SELECT' statement with few key differences. Suppose we want to find out count employee names whos's salary is more than 50000. We can use a 'HAVING' clause with 'SELECT' statement as below.

    -   SELECT COUNT(first_name) , sex FROM employee GROUP BY sex HAVING MIN(salary) > 50000 ;

This showcases that there are 6 male and 3 females present in the data who earns more than 50000. HAVING clause is generally used with GROUP BY statement. This needs to be remembered.
